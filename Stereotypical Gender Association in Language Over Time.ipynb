# Evaluating Gender Bias

```python
import numpy as np
from plotting_utility import evaluate_bias_subject_language
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

topic_bias_year, regression_params = evaluate_bias_subject_language(gender_profile_lang)

topic_bias_year = np.array(topic_bias_year)
hypothetical_score = [0.0 for i in range(len(topic_bias_year))]
year_begin = 1800
year_range = []
cos_dist = []

for k in range(20):
    year_range.append(year_begin)

    decade_score = topic_bias_year[:, k]
    dist = np.average(np.absolute(decade_score))
    
    cos_dist.append(dist)
    year_begin += 10


plt.figure(figsize=(8, 6))
plt.plot(cos_dist, linestyle='-', linewidth=3, alpha=0.75)
plt.plot(cos_dist, marker='o', linestyle='-', linewidth=0.1, \
         ms=6, mfc='blue', mec='black', mew=1.25)
plt.xticks([k for k in range(0, 20)], \
        [1800+x for x in range(0, 200, 10)], rotation=45)
plt.ylim(0.0, 0.06)
plt.title("Evaluation of Shift in Gender Bias")
plt.show()
```
